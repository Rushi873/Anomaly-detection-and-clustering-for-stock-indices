```python
import yfinance as yf
import datetime
import pandas as pd
import numpy as np
```

```python
# Getting the index data from yahoo finance

start_date = "2000-01-01"
end_date = "2024-06-01"
ticker = "^BVSP"

stock_data = yf.download(ticker, start=start_date, end=end_date, interval = '1d')
```
