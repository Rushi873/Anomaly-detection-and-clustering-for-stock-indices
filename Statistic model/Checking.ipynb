```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
NIFTY = pd.read_csv('D:/UFG/Data collection/NIFTY.csv')
NIFTY['Date']= pd.to_datetime(NIFTY['Date'], format='%Y-%m-%d')
```

```python
NIFTY
```

```python
# Filter data 
start_date = pd.to_datetime("2000-01-01", format='%Y-%m-%d')
end_date = pd.to_datetime("2024-06-01", format='%Y-%m-%d')
filtered_data = NIFTY[(NIFTY['Date'] >= start_date) & (NIFTY['Date'] < end_date)]
```

```python
filtered_data
```

```python
plt.figure(figsize=(14, 7))
plt.plot(filtered_data['Date'], filtered_data['Close'], label='NIFTY Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
